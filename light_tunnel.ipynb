```python
import cv2
import numpy as np
from skimage.draw import circle_perimeter_aa, line_aa

cam = cv2.VideoCapture(0)

HEIGHT = 480
WIDTH = 640

ANGLE_TO_CORNER = np.arctan(HEIGHT / WIDTH)


RADIUS = 140

y_center = HEIGHT // 2
x_center = WIDTH // 2

rr, cc, val = circle_perimeter_aa(y_center, x_center, RADIUS)


while True:
    
    ret, frame = cam.read()


    for r, c in zip(rr,cc): # iterate through each pixel on the circle

        pixel_color = frame[r,c]

        angle = np.arctan(np.abs(y_center - r) / (np.abs(x_center - c) + 0.00001))



        if r <= y_center and c >= x_center: # first quadrant

            if  angle <= ANGLE_TO_CORNER: 

                line_y, line_x, _ = line_aa(r, c, 240 - int(320*np.tan(angle)), 639)


            else:

                line_y, line_x, _ = line_aa(r, c, 0, 320 + int(240/np.tan(angle)))



        elif r <= y_center and c <= x_center: # second quadrant

            if angle <= ANGLE_TO_CORNER:

                line_y, line_x, _ = line_aa(r, c, 240 - int(320*np.tan(angle)), 0)
            else:
                line_y, line_x, _ = line_aa(r, c, 0, 320 - int(240/np.tan(angle)))




        elif r >= y_center and c <= x_center: # third quadrant
            if angle <= ANGLE_TO_CORNER:
                line_y, line_x, _ = line_aa(r, c, 240 + int(320*np.tan(angle)), 0)
            else:
                line_y, line_x, _ = line_aa(r, c, 479, 320 - int(240/np.tan(angle)))

        elif r >= y_center and c >= x_center: # fourth quadrant

            if angle <= ANGLE_TO_CORNER:
                line_y, line_x, _ = line_aa(r, c, 240 + int(320*np.tan(angle)), 639)

            else:
                line_y, line_x, _ = line_aa(r, c, 479, 320 + int(240/np.tan(angle)))




        if np.all((0 <= line_y) & (line_y < HEIGHT)) and np.all((0 <= line_x) & (line_x < WIDTH)):
            frame[line_y, line_x, :] = pixel_color
        else:
            print(r,c, line_y, line_x)

        

    
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# After the loop release the cap object
cam.release()
# Destroy all the windows
cv2.destroyAllWindows()   
```
